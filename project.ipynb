```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fire_data = pd.read_csv('fire_2019.csv', encoding='euc-kr')
weather_data = pd.read_csv('weather.csv', encoding='euc-kr')
```

```python
mask = fire_data.시도 == '대전광역시'
fire_data_대전 = fire_data[mask]
```

```python
weather_list = []
fire_dict = {}
fire_dict_all = {}

for i in range(len(weather_data.일시)):
    일시 = weather_data.iloc[i]['일시']
    습도 = weather_data.iloc[i]['평균 상대습도(%)']
    weather_list.append(습도)
    fire_dict[일시[5:10]] = 0
    fire_dict_all[일시[5:10]] = 0

for i in fire_data_대전.화재발생년월일:
    fire_dict[i[5:10]] += 1

for i in fire_data.화재발생년월일:
    fire_dict_all[i[5:10]] += 1
```

```python
plt.rc('font', family='Malgun Gothic')
fig, ax1 = plt.subplots(figsize=(20,5))
plt.title('습도에 따른 화재발생건수 비교 그래프(대전)', size=20)
line1 = ax1.plot(weather_list, color='red', label='상대습도(%)')
ax1.set_xlabel('일시')
ax1.set_ylabel('상대습도(%)')

ax2 = ax1.twinx()
line2 = ax2.bar(fire_dict.keys(), fire_dict.values(), color='orange', label='화재발생건수')
ax2.set_ylabel('화재발생건수(건)')
plt.xticks([0, 31, 59, 90, 120, 151, 181, 212, 243, 273, 304, 334],
           rotation=45)
labels = [line1[0].get_label(), line2.get_label()]

ax1.set_zorder(3)
ax2.set_zorder(2)
ax1.patch.set_visible(False)
ax1.legend([line1[0], line2], labels)
plt.savefig('대전.jpg')
plt.show()
```
